```python
import os

# Memory-safe settings for Windows/Jupyter. These must be set before
# importing NumPy, scikit-learn, Joblib or TensorFlow.
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["OPENBLAS_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["NUMEXPR_NUM_THREADS"] = "1"
os.environ["VECLIB_MAXIMUM_THREADS"] = "1"
os.environ["TF_NUM_INTRAOP_THREADS"] = "1"
os.environ["TF_NUM_INTEROP_THREADS"] = "1"
os.environ["JOBLIB_MULTIPROCESSING"] = "0"
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

import gc
import json
import random
import warnings
from datetime import datetime

import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import yaml

from sklearn.base import clone
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import ExtraTreesRegressor, GradientBoostingRegressor, RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import KFold, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import BatchNormalization, Dense, Dropout, Input
from tensorflow.keras.models import Sequential

warnings.filterwarnings("ignore")

# Limit TensorFlow CPU threads as an additional safeguard.
try:
    tf.config.threading.set_intra_op_parallelism_threads(1)
    tf.config.threading.set_inter_op_parallelism_threads(1)
except RuntimeError:
    pass

RANDOM_STATE = 42
TEST_SIZE = 0.20
TARGET_COLUMN = "11-Sep"
ANN_EPOCHS = 250
ANN_BATCH_SIZE = 8

# CSA settings. Increase these values only if sufficient RAM is available.
CSA_NUMBER_OF_CROWS = 6
CSA_ITERATIONS = 5
CSA_AWARENESS_PROBABILITY = 0.20
CSA_FLIGHT_LENGTH = 2.0

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

PROJECT_FOLDER = (
    r"C:\Users\NXTWAVE\Downloads"
    r"\Smart Rainfall Pattern Analysis and Localized Rainfall Prediction System"
)
DATASET_PATH = os.path.join(PROJECT_FOLDER, "rainfall_0.csv")
os.makedirs(PROJECT_FOLDER, exist_ok=True)

# Every output starts with csa_.
PATHS = {
    "h5": os.path.join(PROJECT_FOLDER, "csa_rainfall_ann_model.h5"),
    "pkl": os.path.join(PROJECT_FOLDER, "csa_rainfall_model.pkl"),
    "preprocessor": os.path.join(PROJECT_FOLDER, "csa_rainfall_preprocessor.pkl"),
    "yaml": os.path.join(PROJECT_FOLDER, "csa_rainfall_config.yaml"),
    "json": os.path.join(PROJECT_FOLDER, "csa_rainfall_results.json"),
    "result_csv": os.path.join(PROJECT_FOLDER, "csa_result.csv"),
    "prediction_csv": os.path.join(PROJECT_FOLDER, "csa_prediction.csv"),
    "optimization_csv": os.path.join(PROJECT_FOLDER, "csa_optimization_history.csv"),
    "comparison_csv": os.path.join(PROJECT_FOLDER, "csa_model_comparison.csv"),
    "importance_csv": os.path.join(PROJECT_FOLDER, "csa_feature_importance.csv"),
    "accuracy_graph": os.path.join(PROJECT_FOLDER, "csa_accuracy_graph.png"),
    "comparison_graph": os.path.join(PROJECT_FOLDER, "csa_comparison_graph.png"),
    "heatmap": os.path.join(PROJECT_FOLDER, "csa_heatmap.png"),
    "result_graph": os.path.join(PROJECT_FOLDER, "csa_result_graph.png"),
    "prediction_graph": os.path.join(PROJECT_FOLDER, "csa_prediction_graph.png"),
    "optimization_graph": os.path.join(PROJECT_FOLDER, "csa_optimization_graph.png"),
    "ann_history_graph": os.path.join(PROJECT_FOLDER, "csa_ann_training_history.png"),
    "importance_graph": os.path.join(PROJECT_FOLDER, "csa_feature_importance.png"),
}

CSA_SEARCH_SPACE = {
    "n_estimators": [100, 150, 200, 250, 300, 350, 400],
    "max_depth": [None, 4, 6, 8, 10, 12, 15, 20],
    "min_samples_split": [2, 3, 4, 5, 6, 8],
    "min_samples_leaf": [1, 2, 3, 4, 5],
    "max_features": ["sqrt", "log2", 0.5, 0.7, 1.0],
    "bootstrap": [True, False],
}
PARAMETER_NAMES = list(CSA_SEARCH_SPACE)
CSA_DIMENSIONS = len(PARAMETER_NAMES)


def create_one_hot_encoder():
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)


def calculate_metrics(actual, predicted):
    actual = np.asarray(actual, dtype=float).reshape(-1)
    predicted = np.asarray(predicted, dtype=float).reshape(-1)
    mse = mean_squared_error(actual, predicted)
    return {
        "MAE": float(mean_absolute_error(actual, predicted)),
        "MSE": float(mse),
        "RMSE": float(np.sqrt(mse)),
        "R2_Score": float(r2_score(actual, predicted)) if len(actual) > 1 else 0.0,
    }


def json_converter(value):
    if isinstance(value, np.integer):
        return int(value)
    if isinstance(value, np.floating):
        value = float(value)
        return None if np.isnan(value) or np.isinf(value) else value
    if isinstance(value, np.ndarray):
        return value.tolist()
    if isinstance(value, pd.Timestamp):
        return value.isoformat()
    raise TypeError(f"Object of type {type(value).__name__} is not JSON serializable")


def save_and_show_graph(path):
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    print("Saved:", path)
    plt.show()
    plt.close()


def create_ann_model(input_dimension):
    model = Sequential([
        Input(shape=(input_dimension,)),
        Dense(128, activation="relu"),
        BatchNormalization(),
        Dropout(0.20),
        Dense(64, activation="relu"),
        BatchNormalization(),
        Dropout(0.15),
        Dense(32, activation="relu"),
        Dropout(0.10),
        Dense(16, activation="relu"),
        Dense(1, activation="linear"),
    ])
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss="mean_squared_error",
        metrics=["mean_absolute_error"],
    )
    return model


def position_to_parameters(position):
    parameters = {}
    for dimension, name in enumerate(PARAMETER_NAMES):
        values = CSA_SEARCH_SPACE[name]
        normalized = float(np.clip(position[dimension], 0.0, 1.0))
        index = min(int(normalized * len(values)), len(values) - 1)
        parameters[name] = values[index]
    return parameters


def parameter_key(parameters):
    return tuple((name, str(parameters[name])) for name in sorted(parameters))


def build_random_forest(parameters):
    # n_jobs=1 is essential here to avoid Windows BrokenProcessPool/MemoryError.
    return RandomForestRegressor(
        n_estimators=parameters["n_estimators"],
        max_depth=parameters["max_depth"],
        min_samples_split=parameters["min_samples_split"],
        min_samples_leaf=parameters["min_samples_leaf"],
        max_features=parameters["max_features"],
        bootstrap=parameters["bootstrap"],
        random_state=RANDOM_STATE,
        n_jobs=1,
    )


def sequential_cv_rmse(model, X_data, y_data, folds=3):
    """Memory-safe cross-validation with no Joblib worker processes."""
    splitter = KFold(n_splits=folds, shuffle=True, random_state=RANDOM_STATE)
    scores = []
    for fold_number, (train_indices, validation_indices) in enumerate(splitter.split(X_data), 1):
        fold_model = clone(model)
        fold_model.fit(X_data[train_indices], y_data[train_indices])
        fold_predictions = np.maximum(fold_model.predict(X_data[validation_indices]), 0)
        fold_rmse = np.sqrt(mean_squared_error(y_data[validation_indices], fold_predictions))
        scores.append(float(fold_rmse))
        del fold_model, fold_predictions
        gc.collect()
    return float(np.mean(scores))


def evaluate_crow(position, X_training, y_training, cache):
    parameters = position_to_parameters(position)
    key = parameter_key(parameters)
    if key in cache:
        return cache[key]
    model = build_random_forest(parameters)
    rmse = sequential_cv_rmse(model, X_training, y_training, folds=3)
    result = {"parameters": parameters, "rmse": rmse, "fitness": -rmse}
    cache[key] = result
    del model
    gc.collect()
    return result


def crow_search_optimization(X_training, y_training):
    evaluation_cache = {}
    positions = np.random.uniform(0.0, 1.0, size=(CSA_NUMBER_OF_CROWS, CSA_DIMENSIONS))
    memories = positions.copy()
    memory_rmse = np.full(CSA_NUMBER_OF_CROWS, np.inf)
    history = []

    global_best_position = None
    global_best_parameters = None
    global_best_rmse = np.inf

    print("\n" + "=" * 78)
    print("MEMORY-SAFE CROW SEARCH ALGORITHM OPTIMIZATION")
    print("=" * 78)

    # Initial evaluation.
    for i in range(CSA_NUMBER_OF_CROWS):
        evaluation = evaluate_crow(positions[i], X_training, y_training, evaluation_cache)
        memory_rmse[i] = evaluation["rmse"]
        if evaluation["rmse"] < global_best_rmse:
            global_best_rmse = evaluation["rmse"]
            global_best_position = positions[i].copy()
            global_best_parameters = evaluation["parameters"].copy()

    for iteration in range(1, CSA_ITERATIONS + 1):
        print(f"\nCSA iteration {iteration}/{CSA_ITERATIONS}")
        new_positions = np.zeros_like(positions)

        for i in range(CSA_NUMBER_OF_CROWS):
            choices = [j for j in range(CSA_NUMBER_OF_CROWS) if j != i]
            followed = random.choice(choices)
            if random.random() >= CSA_AWARENESS_PROBABILITY:
                random_vector = np.random.random(CSA_DIMENSIONS)
                new_position = (
                    positions[i]
                    + CSA_FLIGHT_LENGTH
                    * random_vector
                    * (memories[followed] - positions[i])
                )
            else:
                new_position = np.random.uniform(0.0, 1.0, size=CSA_DIMENSIONS)
            new_positions[i] = np.clip(new_position, 0.0, 1.0)

        iteration_scores = []
        for i in range(CSA_NUMBER_OF_CROWS):
            evaluation = evaluate_crow(new_positions[i], X_training, y_training, evaluation_cache)
            current_rmse = evaluation["rmse"]
            iteration_scores.append(current_rmse)
            print(f"  Crow {i + 1}: CV RMSE = {current_rmse:.4f}")

            if current_rmse < memory_rmse[i]:
                memories[i] = new_positions[i].copy()
                memory_rmse[i] = current_rmse

            if current_rmse < global_best_rmse:
                global_best_rmse = current_rmse
                global_best_position = new_positions[i].copy()
                global_best_parameters = evaluation["parameters"].copy()

        positions = new_positions
        history.append({
            "Iteration": iteration,
            "Iteration_Best_CV_RMSE": float(np.min(iteration_scores)),
            "Average_CV_RMSE": float(np.mean(iteration_scores)),
            "Global_Best_CV_RMSE": float(global_best_rmse),
            "Best_n_estimators": global_best_parameters["n_estimators"],
            "Best_max_depth": str(global_best_parameters["max_depth"]),
            "Best_min_samples_split": global_best_parameters["min_samples_split"],
            "Best_min_samples_leaf": global_best_parameters["min_samples_leaf"],
            "Best_max_features": str(global_best_parameters["max_features"]),
            "Best_bootstrap": global_best_parameters["bootstrap"],
        })
        print("  Global best CV RMSE:", round(global_best_rmse, 4))
        print("  Global best parameters:", global_best_parameters)

    return (
        global_best_parameters,
        float(global_best_rmse),
        global_best_position,
        pd.DataFrame(history),
    )


print("=" * 78)
print("SMART RAINFALL PREDICTION USING CROW SEARCH ALGORITHM")
print("=" * 78)
print("Project folder:", PROJECT_FOLDER)
print("Dataset path:", DATASET_PATH)

if not os.path.exists(DATASET_PATH):
    raise FileNotFoundError(
        f"Dataset not found:\n{DATASET_PATH}\nPlace rainfall_0.csv in the project folder."
    )

data = pd.read_csv(DATASET_PATH)
data.columns = data.columns.astype(str).str.strip()
print("\nDataset loaded successfully.")
print("Dataset shape:", data.shape)
print(data.head())

if "S.No" in data.columns:
    data = data.drop(columns=["S.No"])

if TARGET_COLUMN not in data.columns:
    raise ValueError(
        f"Target column '{TARGET_COLUMN}' was not found. Available columns: {data.columns.tolist()}"
    )

categorical_columns = [c for c in ["District", "Taluka", "Circle"] if c in data.columns]
rainfall_columns = [c for c in data.columns if c not in categorical_columns and c != TARGET_COLUMN]
feature_columns = categorical_columns + rainfall_columns

for column in categorical_columns:
    data[column] = data[column].fillna("Unknown").astype(str).str.strip()
    data[column] = data[column].replace({"": "Unknown", "nan": "Unknown", "None": "Unknown"})

for column in rainfall_columns + [TARGET_COLUMN]:
    data[column] = pd.to_numeric(data[column], errors="coerce")

data = data.dropna(subset=[TARGET_COLUMN]).reset_index(drop=True)
if len(data) < 10:
    raise ValueError("Too few usable rows are available for model training.")

X = data[feature_columns].copy()
y = data[TARGET_COLUMN].astype(float).copy()
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE
)

numeric_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", create_one_hot_encoder()),
])
preprocessor = ColumnTransformer([
    ("numeric", numeric_pipeline, rainfall_columns),
    ("categorical", categorical_pipeline, categorical_columns),
], remainder="drop")

X_train_processed = np.asarray(preprocessor.fit_transform(X_train), dtype=np.float32)
X_test_processed = np.asarray(preprocessor.transform(X_test), dtype=np.float32)
X_all_processed = np.asarray(preprocessor.transform(X), dtype=np.float32)
y_train_array = np.asarray(y_train, dtype=np.float32).reshape(-1)
y_test_array = np.asarray(y_test, dtype=np.float32).reshape(-1)

print("Training records:", len(X_train))
print("Testing records:", len(X_test))
print("Processed training shape:", X_train_processed.shape)

joblib.dump(preprocessor, PATHS["preprocessor"], compress=3)
print("Preprocessor saved:", PATHS["preprocessor"])

best_parameters, best_cv_rmse, best_position, optimization_history = crow_search_optimization(
    X_train_processed, y_train_array
)
optimization_history.to_csv(PATHS["optimization_csv"], index=False)

csa_model = build_random_forest(best_parameters)
csa_model.fit(X_train_processed, y_train_array)
csa_predictions = np.maximum(csa_model.predict(X_test_processed), 0)
csa_metrics = calculate_metrics(y_test_array, csa_predictions)

print("\nCSA model metrics:")
for name, value in csa_metrics.items():
    print(f"  {name}: {value:.4f}")

comparison_models = {
    "Linear Regression": LinearRegression(),
    "Standard Random Forest": RandomForestRegressor(
        n_estimators=250, max_depth=10, max_features="sqrt",
        random_state=RANDOM_STATE, n_jobs=1
    ),
    "Gradient Boosting": GradientBoostingRegressor(
        n_estimators=250, learning_rate=0.03, max_depth=3, random_state=RANDOM_STATE
    ),
    "Extra Trees": ExtraTreesRegressor(
        n_estimators=300, max_depth=10, max_features="sqrt",
        random_state=RANDOM_STATE, n_jobs=1
    ),
}
model_metrics = {"CSA Random Forest": csa_metrics}
model_predictions = {"CSA Random Forest": csa_predictions}

for model_name, model in comparison_models.items():
    print("Training", model_name)
    model.fit(X_train_processed, y_train_array)
    predictions = np.maximum(model.predict(X_test_processed), 0)
    model_predictions[model_name] = predictions
    model_metrics[model_name] = calculate_metrics(y_test_array, predictions)
    del model
    gc.collect()

# ANN is trained to generate the requested H5 model.
ann_model = create_ann_model(X_train_processed.shape[1])
early_stopping = EarlyStopping(monitor="val_loss", patience=30, restore_best_weights=True, verbose=1)
reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=10, min_lr=0.00001, verbose=1)
ann_history = ann_model.fit(
    X_train_processed,
    y_train_array,
    validation_split=0.20,
    epochs=ANN_EPOCHS,
    batch_size=ANN_BATCH_SIZE,
    callbacks=[early_stopping, reduce_lr],
    verbose=1,
)
ann_predictions = np.maximum(ann_model.predict(X_test_processed, verbose=0).reshape(-1), 0)
model_predictions["ANN"] = ann_predictions
model_metrics["ANN"] = calculate_metrics(y_test_array, ann_predictions)
ann_model.save(PATHS["h5"])

model_package = {
    "project_name": "Smart Rainfall Pattern Analysis and Localized Rainfall Prediction System",
    "optimization_method": "Crow Search Algorithm",
    "model_name": "CSA-Optimized Random Forest Regressor",
    "model": csa_model,
    "preprocessor": preprocessor,
    "feature_columns": feature_columns,
    "categorical_columns": categorical_columns,
    "rainfall_columns": rainfall_columns,
    "target_column": TARGET_COLUMN,
    "best_parameters": best_parameters,
    "best_position": best_position.tolist(),
    "cross_validation_rmse": best_cv_rmse,
    "test_metrics": csa_metrics,
    "created_at": datetime.now().isoformat(),
}
joblib.dump(model_package, PATHS["pkl"], compress=3)

# Test-set result CSV.
result_data = X_test[categorical_columns].copy().reset_index(drop=True)
result_data["Actual_Rainfall_11_Sep"] = np.round(y_test_array, 4)
result_data["CSA_Predicted_Rainfall_11_Sep"] = np.round(csa_predictions, 4)
result_data["Absolute_Error"] = np.round(np.abs(y_test_array - csa_predictions), 4)
result_data["Squared_Error"] = np.round((y_test_array - csa_predictions) ** 2, 4)
result_data["Prediction_Difference"] = np.round(csa_predictions - y_test_array, 4)
result_data["Model_Used"] = "CSA-Optimized Random Forest"
result_data.to_csv(PATHS["result_csv"], index=False)

# All-location prediction CSV.
all_predictions = np.maximum(csa_model.predict(X_all_processed), 0)
prediction_data = data[categorical_columns].copy().reset_index(drop=True)
prediction_data["Actual_Rainfall_11_Sep"] = np.round(y.to_numpy(), 4)
prediction_data["CSA_Predicted_Rainfall_11_Sep"] = np.round(all_predictions, 4)
prediction_data["Absolute_Error"] = np.round(np.abs(y.to_numpy() - all_predictions), 4)
prediction_data["Rainfall_Category"] = np.select(
    [all_predictions >= 64.5, all_predictions >= 15.6, all_predictions >= 2.5],
    ["Heavy Rainfall", "Moderate Rainfall", "Light Rainfall"],
    default="Very Light or No Rainfall",
)
prediction_data["Model_Used"] = "CSA-Optimized Random Forest"
prediction_data.to_csv(PATHS["prediction_csv"], index=False)

comparison_data = pd.DataFrame([
    {"Model": name, **metrics} for name, metrics in model_metrics.items()
]).sort_values("RMSE").reset_index(drop=True)
comparison_data.to_csv(PATHS["comparison_csv"], index=False)

# Accuracy graph using R².
model_names = list(model_metrics)
r2_scores = [model_metrics[name]["R2_Score"] for name in model_names]
plt.figure(figsize=(14, 7))
bars = plt.bar(model_names, r2_scores)
plt.axhline(0, linewidth=1)
plt.title("CSA Rainfall Prediction Model Accuracy", fontsize=16, fontweight="bold")
plt.xlabel("Regression models")
plt.ylabel("R² score")
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.4)
for bar, score in zip(bars, r2_scores):
    plt.text(bar.get_x() + bar.get_width() / 2, score, f"{score:.3f}", ha="center", va="bottom" if score >= 0 else "top")
save_and_show_graph(PATHS["accuracy_graph"])

# MAE and RMSE comparison graph.
mae_values = [model_metrics[name]["MAE"] for name in model_names]
rmse_values = [model_metrics[name]["RMSE"] for name in model_names]
positions = np.arange(len(model_names))
width = 0.35
plt.figure(figsize=(14, 7))
plt.bar(positions - width / 2, mae_values, width=width, label="MAE")
plt.bar(positions + width / 2, rmse_values, width=width, label="RMSE")
plt.title("CSA Model Error Comparison", fontsize=16, fontweight="bold")
plt.xlabel("Regression models")
plt.ylabel("Error value")
plt.xticks(positions, model_names, rotation=20)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["comparison_graph"])

# Rainfall correlation heatmap.
correlation_matrix = data[rainfall_columns + [TARGET_COLUMN]].corr()
plt.figure(figsize=(20, 16))
image = plt.imshow(correlation_matrix, aspect="auto", cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(image, label="Correlation coefficient")
plt.title("CSA Rainfall Correlation Heatmap", fontsize=18, fontweight="bold")
plt.xticks(np.arange(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90, fontsize=7)
plt.yticks(np.arange(len(correlation_matrix.index)), correlation_matrix.index, fontsize=7)
save_and_show_graph(PATHS["heatmap"])

# Test result graph.
sorted_result = result_data.sort_values("Actual_Rainfall_11_Sep").reset_index(drop=True)
samples = np.arange(1, len(sorted_result) + 1)
plt.figure(figsize=(14, 7))
plt.plot(samples, sorted_result["Actual_Rainfall_11_Sep"], marker="o", label="Actual rainfall")
plt.plot(samples, sorted_result["CSA_Predicted_Rainfall_11_Sep"], marker="s", label="CSA predicted rainfall")
plt.title("CSA Actual versus Predicted Rainfall Results", fontsize=16, fontweight="bold")
plt.xlabel("Test sample number")
plt.ylabel("Rainfall in millimetres")
plt.legend()
plt.grid(linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["result_graph"])

# All-location prediction graph.
sorted_prediction = prediction_data.sort_values("Actual_Rainfall_11_Sep").reset_index(drop=True)
locations = np.arange(1, len(sorted_prediction) + 1)
plt.figure(figsize=(16, 8))
plt.plot(locations, sorted_prediction["Actual_Rainfall_11_Sep"], marker="o", markersize=4, label="Actual rainfall")
plt.plot(locations, sorted_prediction["CSA_Predicted_Rainfall_11_Sep"], marker="s", markersize=4, label="CSA predicted rainfall")
plt.title("CSA Localized Rainfall Prediction for All Locations", fontsize=16, fontweight="bold")
plt.xlabel("Location number")
plt.ylabel("Rainfall in millimetres")
plt.legend()
plt.grid(linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["prediction_graph"])

# CSA convergence graph.
plt.figure(figsize=(12, 7))
plt.plot(optimization_history["Iteration"], optimization_history["Global_Best_CV_RMSE"], marker="o", label="Global best CV RMSE")
plt.plot(optimization_history["Iteration"], optimization_history["Average_CV_RMSE"], marker="s", label="Average crow CV RMSE")
plt.plot(optimization_history["Iteration"], optimization_history["Iteration_Best_CV_RMSE"], marker="^", label="Iteration best CV RMSE")
plt.title("Crow Search Algorithm Optimization Progress", fontsize=16, fontweight="bold")
plt.xlabel("CSA iteration")
plt.ylabel("Cross-validation RMSE")
plt.xticks(optimization_history["Iteration"])
plt.legend()
plt.grid(linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["optimization_graph"])

# ANN history graph.
training_loss = ann_history.history.get("loss", [])
validation_loss = ann_history.history.get("val_loss", [])
plt.figure(figsize=(12, 7))
plt.plot(range(1, len(training_loss) + 1), training_loss, label="Training loss")
plt.plot(range(1, len(validation_loss) + 1), validation_loss, label="Validation loss")
plt.title("CSA Project ANN Training History", fontsize=16, fontweight="bold")
plt.xlabel("Epoch")
plt.ylabel("Mean squared error loss")
plt.legend()
plt.grid(linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["ann_history_graph"])

# Feature importance CSV and graph.
try:
    feature_names = preprocessor.get_feature_names_out()
except Exception:
    feature_names = np.array([f"Feature_{i + 1}" for i in range(X_train_processed.shape[1])])
importance_data = pd.DataFrame({"Feature": feature_names, "Importance": csa_model.feature_importances_})
importance_data = importance_data.sort_values("Importance", ascending=False).reset_index(drop=True)
importance_data.to_csv(PATHS["importance_csv"], index=False)
top_importance = importance_data.head(20).sort_values("Importance")
plt.figure(figsize=(12, 9))
plt.barh(top_importance["Feature"], top_importance["Importance"])
plt.title("Top CSA Random Forest Feature Importances", fontsize=16, fontweight="bold")
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.grid(axis="x", linestyle="--", alpha=0.4)
save_and_show_graph(PATHS["importance_graph"])

configuration = {
    "project": {
        "name": "Smart Rainfall Pattern Analysis and Localized Rainfall Prediction System",
        "optimization_method": "Crow Search Algorithm",
        "created_at": datetime.now().isoformat(),
    },
    "dataset": {
        "path": DATASET_PATH,
        "rows": int(data.shape[0]),
        "columns": int(data.shape[1]),
        "training_rows": int(len(X_train)),
        "testing_rows": int(len(X_test)),
        "target_column": TARGET_COLUMN,
    },
    "features": {
        "categorical_columns": categorical_columns,
        "rainfall_columns": rainfall_columns,
        "processed_feature_count": int(X_train_processed.shape[1]),
    },
    "csa_settings": {
        "number_of_crows": CSA_NUMBER_OF_CROWS,
        "iterations": CSA_ITERATIONS,
        "awareness_probability": CSA_AWARENESS_PROBABILITY,
        "flight_length": CSA_FLIGHT_LENGTH,
        "memory_safe_n_jobs": 1,
    },
    "best_csa_parameters": best_parameters,
    "best_csa_position": best_position.tolist(),
    "best_cross_validation_rmse": best_cv_rmse,
    "csa_test_metrics": csa_metrics,
    "all_model_metrics": model_metrics,
    "saved_files": PATHS,
}

with open(PATHS["yaml"], "w", encoding="utf-8") as file:
    yaml.safe_dump(configuration, file, sort_keys=False, allow_unicode=True)

json_results = {
    **configuration,
    "target_statistics": {
        "minimum": float(y.min()),
        "maximum": float(y.max()),
        "mean": float(y.mean()),
        "median": float(y.median()),
        "standard_deviation": float(y.std()),
    },
    "ann_training": {
        "epochs_completed": int(len(training_loss)),
        "final_training_loss": float(training_loss[-1]) if training_loss else None,
        "final_validation_loss": float(validation_loss[-1]) if validation_loss else None,
    },
}
with open(PATHS["json"], "w", encoding="utf-8") as file:
    json.dump(json_results, file, indent=4, ensure_ascii=False, default=json_converter)

# Verify the saved model package.
loaded_package = joblib.load(PATHS["pkl"])
sample_input = X.iloc[[0]].copy()
sample_processed = np.asarray(loaded_package["preprocessor"].transform(sample_input), dtype=np.float32)
sample_prediction = max(float(loaded_package["model"].predict(sample_processed)[0]), 0)
print("\nSample prediction:", round(sample_prediction, 2), "mm")
print("Sample actual value:", round(float(y.iloc[0]), 2), "mm")

print("\n" + "=" * 78)
print("CSA PROJECT COMPLETED SUCCESSFULLY")
print("=" * 78)
for name, path in PATHS.items():
    status = "CREATED" if os.path.exists(path) else "NOT CREATED"
    print(f"[{status}] {os.path.basename(path)}")
print("\nBest CSA parameters:", best_parameters)
print("CSA RMSE:", round(csa_metrics["RMSE"], 4))
print("CSA R² Score:", round(csa_metrics["R2_Score"], 4))
print("\nAll CSA results were saved in:", PROJECT_FOLDER)
```

```python
import os

# Limit native-library and TensorFlow threads before importing NumPy,
# scikit-learn or TensorFlow. This prevents Windows MemoryError and
# BrokenProcessPool failures caused by nested parallel processing.
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["OPENBLAS_NUM_THREADS"] = "1"
os.environ["MKL_NUM_THREADS"] = "1"
os.environ["NUMEXPR_NUM_THREADS"] = "1"
os.environ["VECLIB_MAXIMUM_THREADS"] = "1"
os.environ["TF_NUM_INTRAOP_THREADS"] = "1"
os.environ["TF_NUM_INTEROP_THREADS"] = "1"
os.environ["JOBLIB_MULTIPROCESSING"] = "0"

import json
import random
import warnings
from datetime import datetime

import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import yaml

from sklearn.compose import ColumnTransformer
from sklearn.ensemble import (
    ExtraTreesRegressor,
    GradientBoostingRegressor,
    RandomForestRegressor
)
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.model_selection import (
    KFold,
    cross_val_score,
    train_test_split
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from tensorflow.keras.callbacks import (
    EarlyStopping,
    ReduceLROnPlateau
)
from tensorflow.keras.layers import (
    BatchNormalization,
    Dense,
    Dropout,
    Input
)
from tensorflow.keras.models import Sequential


# =====================================================================
# GENERAL SETTINGS
# =====================================================================

warnings.filterwarnings("ignore")

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

RANDOM_STATE = 42
TEST_SIZE = 0.20

random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

try:
    tf.config.threading.set_intra_op_parallelism_threads(1)
    tf.config.threading.set_inter_op_parallelism_threads(1)
except RuntimeError:
    # TensorFlow may already be initialized in a reused Jupyter kernel.
    pass


# =====================================================================
# PROJECT PATH
# =====================================================================

PROJECT_FOLDER = (
    r"C:\Users\NXTWAVE\Downloads"
    r"\Smart Rainfall Pattern Analysis and Localized Rainfall Prediction System"
)

DATASET_PATH = os.path.join(
    PROJECT_FOLDER,
    "rainfall_0.csv"
)

os.makedirs(PROJECT_FOLDER, exist_ok=True)


# =====================================================================
# AIS OUTPUT FILE PATHS
# =====================================================================

AIS_H5_MODEL_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_rainfall_ann_model.h5"
)

AIS_PKL_MODEL_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_rainfall_model.pkl"
)

AIS_PREPROCESSOR_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_rainfall_preprocessor.pkl"
)

AIS_YAML_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_rainfall_config.yaml"
)

AIS_JSON_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_rainfall_results.json"
)

AIS_RESULT_CSV_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_result.csv"
)

AIS_PREDICTION_CSV_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_prediction.csv"
)

AIS_OPTIMIZATION_CSV_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_optimization_history.csv"
)

AIS_ACCURACY_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_accuracy_graph.png"
)

AIS_COMPARISON_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_comparison_graph.png"
)

AIS_HEATMAP_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_heatmap.png"
)

AIS_RESULT_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_result_graph.png"
)

AIS_PREDICTION_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_prediction_graph.png"
)

AIS_OPTIMIZATION_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_optimization_graph.png"
)

AIS_ANN_HISTORY_GRAPH_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_ann_training_history.png"
)

AIS_FEATURE_IMPORTANCE_PATH = os.path.join(
    PROJECT_FOLDER,
    "ais_feature_importance.png"
)


# =====================================================================
# TRAINING SETTINGS
# =====================================================================

TARGET_COLUMN = "11-Sep"

ANN_EPOCHS = 300
ANN_BATCH_SIZE = 8

AIS_POPULATION_SIZE = 6
AIS_GENERATIONS = 5
AIS_SELECTED_ANTIBODIES = 3
AIS_CLONES_PER_ANTIBODY = 2
AIS_RANDOM_REPLACEMENTS = 2


# =====================================================================
# HELPER FUNCTION: ONE-HOT ENCODER
# =====================================================================

def create_one_hot_encoder():
    """
    Create a OneHotEncoder compatible with different
    scikit-learn versions.
    """

    try:
        return OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=False
        )

    except TypeError:
        return OneHotEncoder(
            handle_unknown="ignore",
            sparse=False
        )


# =====================================================================
# HELPER FUNCTION: METRICS
# =====================================================================

def calculate_metrics(actual_values, predicted_values):
    """
    Calculate standard regression metrics.
    """

    actual_values = np.asarray(
        actual_values,
        dtype=float
    ).reshape(-1)

    predicted_values = np.asarray(
        predicted_values,
        dtype=float
    ).reshape(-1)

    mae = mean_absolute_error(
        actual_values,
        predicted_values
    )

    mse = mean_squared_error(
        actual_values,
        predicted_values
    )

    rmse = np.sqrt(mse)

    if len(actual_values) > 1:
        r2 = r2_score(
            actual_values,
            predicted_values
        )
    else:
        r2 = 0.0

    return {
        "MAE": float(mae),
        "MSE": float(mse),
        "RMSE": float(rmse),
        "R2_Score": float(r2)
    }


# =====================================================================
# HELPER FUNCTION: JSON CONVERSION
# =====================================================================

def json_converter(value):
    """
    Convert NumPy and Pandas values into normal Python values.
    """

    if isinstance(value, np.integer):
        return int(value)

    if isinstance(value, np.floating):

        converted_value = float(value)

        if np.isnan(converted_value):
            return None

        if np.isinf(converted_value):
            return None

        return converted_value

    if isinstance(value, np.ndarray):
        return value.tolist()

    if isinstance(value, pd.Timestamp):
        return value.isoformat()

    return value


# =====================================================================
# HELPER FUNCTION: SAVE AND SHOW GRAPH
# =====================================================================

def save_and_show_graph(file_path):
    """
    Save the current Matplotlib graph and display it.
    """

    plt.tight_layout()

    plt.savefig(
        file_path,
        dpi=300,
        bbox_inches="tight"
    )

    print("Saved:", file_path)

    plt.show()

    plt.close()


# =====================================================================
# CREATE ANN MODEL
# =====================================================================

def create_ann_model(input_dimension):
    """
    Create an Artificial Neural Network regression model.
    """

    model = Sequential(
        [
            Input(shape=(input_dimension,)),

            Dense(
                128,
                activation="relu"
            ),

            BatchNormalization(),

            Dropout(0.20),

            Dense(
                64,
                activation="relu"
            ),

            BatchNormalization(),

            Dropout(0.15),

            Dense(
                32,
                activation="relu"
            ),

            Dropout(0.10),

            Dense(
                16,
                activation="relu"
            ),

            Dense(
                1,
                activation="linear"
            )
        ]
    )

    optimizer = tf.keras.optimizers.Adam(
        learning_rate=0.001
    )

    model.compile(
        optimizer=optimizer,
        loss="mean_squared_error",
        metrics=["mean_absolute_error"]
    )

    return model


# =====================================================================
# AIS SEARCH SPACE
# =====================================================================

AIS_SEARCH_SPACE = {
    "n_estimators": [
        100,
        150,
        200,
        250,
        300,
        350,
        400,
        500
    ],

    "max_depth": [
        None,
        4,
        6,
        8,
        10,
        12,
        15,
        20
    ],

    "min_samples_split": [
        2,
        3,
        4,
        5,
        6,
        8
    ],

    "min_samples_leaf": [
        1,
        2,
        3,
        4,
        5
    ],

    "max_features": [
        "sqrt",
        "log2",
        0.5,
        0.7,
        1.0
    ],

    "bootstrap": [
        True,
        False
    ]
}


# =====================================================================
# GENERATE RANDOM AIS ANTIBODY
# =====================================================================

def generate_random_antibody():
    """
    Generate one random candidate Random Forest configuration.

    In AIS terminology, each candidate solution is an antibody.
    """

    antibody = {}

    for parameter_name, possible_values in AIS_SEARCH_SPACE.items():

        antibody[parameter_name] = random.choice(
            possible_values
        )

    return antibody


# =====================================================================
# CREATE ANTIBODY KEY
# =====================================================================

def antibody_key(antibody):
    """
    Convert an antibody dictionary into a hashable key.

    This helps avoid evaluating the same solution repeatedly.
    """

    return tuple(
        (
            parameter_name,
            str(antibody[parameter_name])
        )
        for parameter_name in sorted(antibody.keys())
    )


# =====================================================================
# MUTATE AIS ANTIBODY
# =====================================================================

def mutate_antibody(antibody, mutation_rate):
    """
    Mutate an antibody.

    A lower-fitness antibody receives a higher mutation rate.
    """

    mutated_antibody = antibody.copy()

    for parameter_name, possible_values in AIS_SEARCH_SPACE.items():

        if random.random() < mutation_rate:

            current_value = mutated_antibody[
                parameter_name
            ]

            alternative_values = [
                value
                for value in possible_values
                if value != current_value
            ]

            if alternative_values:

                mutated_antibody[
                    parameter_name
                ] = random.choice(
                    alternative_values
                )

    return mutated_antibody


# =====================================================================
# EVALUATE AIS ANTIBODY
# =====================================================================

def evaluate_antibody(
    antibody,
    X_training_data,
    y_training_data,
    evaluation_cache
):
    """
    Evaluate an antibody using cross-validation.

    Fitness is negative RMSE.

    A higher fitness means a lower cross-validation RMSE.
    """

    key = antibody_key(antibody)

    if key in evaluation_cache:
        return evaluation_cache[key]

    model = RandomForestRegressor(
        n_estimators=antibody["n_estimators"],
        max_depth=antibody["max_depth"],
        min_samples_split=antibody["min_samples_split"],
        min_samples_leaf=antibody["min_samples_leaf"],
        max_features=antibody["max_features"],
        bootstrap=antibody["bootstrap"],
        random_state=RANDOM_STATE,
        n_jobs=1
    )

    cross_validation = KFold(
        n_splits=3,
        shuffle=True,
        random_state=RANDOM_STATE
    )

    # Run folds sequentially instead of using joblib multiprocessing.
    # This avoids BrokenProcessPool and MemoryError on Windows/Jupyter.
    rmse_scores = []

    for train_indices, validation_indices in cross_validation.split(
        X_training_data
    ):
        fold_model = RandomForestRegressor(
            n_estimators=antibody["n_estimators"],
            max_depth=antibody["max_depth"],
            min_samples_split=antibody["min_samples_split"],
            min_samples_leaf=antibody["min_samples_leaf"],
            max_features=antibody["max_features"],
            bootstrap=antibody["bootstrap"],
            random_state=RANDOM_STATE,
            n_jobs=1
        )

        fold_model.fit(
            X_training_data[train_indices],
            y_training_data[train_indices]
        )

        fold_predictions = fold_model.predict(
            X_training_data[validation_indices]
        )

        fold_rmse = np.sqrt(
            mean_squared_error(
                y_training_data[validation_indices],
                fold_predictions
            )
        )

        rmse_scores.append(float(fold_rmse))

        del fold_model

    average_rmse = float(
        np.mean(rmse_scores)
    )

    fitness = -average_rmse

    result = {
        "fitness": float(fitness),
        "cv_rmse": float(average_rmse)
    }

    evaluation_cache[key] = result

    return result


# =====================================================================
# ARTIFICIAL IMMUNE SYSTEM OPTIMIZATION
# =====================================================================

def artificial_immune_system_optimization(
    X_training_data,
    y_training_data
):
    """
    Optimize Random Forest hyperparameters with a
    Clonal Selection Artificial Immune System.

    Process:

    1. Generate an initial antibody population.
    2. Measure each antibody's affinity.
    3. Select antibodies with the highest affinity.
    4. Clone the selected antibodies.
    5. Apply hypermutation.
    6. Preserve the best antibodies.
    7. Add new random antibodies.
    8. Repeat for multiple generations.
    """

    evaluation_cache = {}

    population = [
        generate_random_antibody()
        for _ in range(AIS_POPULATION_SIZE)
    ]

    optimization_history = []

    global_best_antibody = None
    global_best_fitness = float("-inf")
    global_best_rmse = float("inf")

    print("\n" + "=" * 75)
    print("ARTIFICIAL IMMUNE SYSTEM OPTIMIZATION")
    print("=" * 75)

    for generation in range(
        1,
        AIS_GENERATIONS + 1
    ):

        print(
            f"\nAIS generation "
            f"{generation}/{AIS_GENERATIONS}"
        )

        evaluated_population = []

        for antibody_number, antibody in enumerate(
            population,
            start=1
        ):

            evaluation = evaluate_antibody(
                antibody,
                X_training_data,
                y_training_data,
                evaluation_cache
            )

            evaluated_population.append(
                {
                    "antibody": antibody,
                    "fitness": evaluation["fitness"],
                    "cv_rmse": evaluation["cv_rmse"]
                }
            )

            print(
                f"  Antibody {antibody_number}: "
                f"CV RMSE = "
                f"{evaluation['cv_rmse']:.4f}"
            )

        evaluated_population = sorted(
            evaluated_population,
            key=lambda item: item["fitness"],
            reverse=True
        )

        generation_best = evaluated_population[0]

        if (
            generation_best["fitness"]
            > global_best_fitness
        ):

            global_best_fitness = generation_best[
                "fitness"
            ]

            global_best_rmse = generation_best[
                "cv_rmse"
            ]

            global_best_antibody = generation_best[
                "antibody"
            ].copy()

        average_generation_rmse = float(
            np.mean(
                [
                    item["cv_rmse"]
                    for item in evaluated_population
                ]
            )
        )

        optimization_history.append(
            {
                "Generation": generation,
                "Best_CV_RMSE": global_best_rmse,
                "Generation_Best_RMSE": generation_best[
                    "cv_rmse"
                ],
                "Average_CV_RMSE": average_generation_rmse,
                "Best_n_estimators": global_best_antibody[
                    "n_estimators"
                ],
                "Best_max_depth": (
                    str(
                        global_best_antibody[
                            "max_depth"
                        ]
                    )
                ),
                "Best_min_samples_split": (
                    global_best_antibody[
                        "min_samples_split"
                    ]
                ),
                "Best_min_samples_leaf": (
                    global_best_antibody[
                        "min_samples_leaf"
                    ]
                ),
                "Best_max_features": (
                    str(
                        global_best_antibody[
                            "max_features"
                        ]
                    )
                ),
                "Best_bootstrap": (
                    global_best_antibody[
                        "bootstrap"
                    ]
                )
            }
        )

        print(
            "  Generation best CV RMSE:",
            round(
                generation_best["cv_rmse"],
                4
            )
        )

        print(
            "  Global best CV RMSE:",
            round(
                global_best_rmse,
                4
            )
        )

        print(
            "  Best antibody:",
            global_best_antibody
        )

        selected_antibodies = evaluated_population[
            :AIS_SELECTED_ANTIBODIES
        ]

        cloned_population = []

        number_of_selected = len(
            selected_antibodies
        )

        for rank, selected_item in enumerate(
            selected_antibodies
        ):

            selected_antibody = selected_item[
                "antibody"
            ]

            rank_fraction = (
                rank + 1
            ) / max(
                number_of_selected,
                1
            )

            mutation_rate = min(
                0.80,
                0.15 + 0.45 * rank_fraction
            )

            for _ in range(
                AIS_CLONES_PER_ANTIBODY
            ):

                cloned_antibody = mutate_antibody(
                    selected_antibody,
                    mutation_rate
                )

                cloned_population.append(
                    cloned_antibody
                )

        cloned_evaluations = []

        for clone in cloned_population:

            evaluation = evaluate_antibody(
                clone,
                X_training_data,
                y_training_data,
                evaluation_cache
            )

            cloned_evaluations.append(
                {
                    "antibody": clone,
                    "fitness": evaluation["fitness"],
                    "cv_rmse": evaluation["cv_rmse"]
                }
            )

        combined_population = (
            evaluated_population
            + cloned_evaluations
        )

        combined_population = sorted(
            combined_population,
            key=lambda item: item["fitness"],
            reverse=True
        )

        retained_count = max(
            AIS_POPULATION_SIZE
            - AIS_RANDOM_REPLACEMENTS,
            1
        )

        next_population = [
            item["antibody"].copy()
            for item in combined_population[
                :retained_count
            ]
        ]

        while len(next_population) < AIS_POPULATION_SIZE:

            next_population.append(
                generate_random_antibody()
            )

        population = next_population

    history_dataframe = pd.DataFrame(
        optimization_history
    )

    return (
        global_best_antibody,
        global_best_rmse,
        history_dataframe
    )


# =====================================================================
# PROGRAM START
# =====================================================================

print("=" * 75)
print("SMART RAINFALL PREDICTION USING AIS - MEMORY SAFE VERSION")
print("=" * 75)

print("\nProject folder:")
print(PROJECT_FOLDER)

print("\nDataset path:")
print(DATASET_PATH)


# =====================================================================
# CHECK DATASET
# =====================================================================

if not os.path.exists(DATASET_PATH):

    raise FileNotFoundError(
        "\nDataset not found.\n"
        f"Expected dataset:\n{DATASET_PATH}\n\n"
        "Place rainfall_0.csv in the project folder."
    )


# =====================================================================
# LOAD DATASET
# =====================================================================

data = pd.read_csv(
    DATASET_PATH
)

data.columns = (
    data.columns
    .astype(str)
    .str.strip()
)

print("\nDataset loaded successfully.")
print("Dataset shape:", data.shape)

print("\nFirst five rows:")
print(data.head())

print("\nDataset columns:")
print(data.columns.tolist())


# =====================================================================
# REMOVE UNNECESSARY INDEX COLUMN
# =====================================================================

if "S.No" in data.columns:

    data = data.drop(
        columns=["S.No"]
    )

    print("\nRemoved unnecessary column: S.No")


# =====================================================================
# IDENTIFY FEATURES
# =====================================================================

if TARGET_COLUMN not in data.columns:

    raise ValueError(
        f"\nTarget column '{TARGET_COLUMN}' was not found.\n"
        f"Available columns:\n{data.columns.tolist()}"
    )

categorical_columns = [
    column
    for column in [
        "District",
        "Taluka",
        "Circle"
    ]
    if column in data.columns
]

rainfall_columns = [
    column
    for column in data.columns
    if column not in categorical_columns
    and column != TARGET_COLUMN
]

feature_columns = (
    categorical_columns
    + rainfall_columns
)

print("\nCategorical columns:")
print(categorical_columns)

print("\nRainfall input columns:")
print(rainfall_columns)

print("\nTarget column:")
print(TARGET_COLUMN)


# =====================================================================
# CLEAN CATEGORICAL DATA
# =====================================================================

for column in categorical_columns:

    data[column] = (
        data[column]
        .fillna("Unknown")
        .astype(str)
        .str.strip()
    )

    data[column] = data[column].replace(
        {
            "": "Unknown",
            "nan": "Unknown",
            "None": "Unknown"
        }
    )


# =====================================================================
# CLEAN NUMERIC RAINFALL DATA
# =====================================================================

for column in rainfall_columns + [TARGET_COLUMN]:

    data[column] = pd.to_numeric(
        data[column],
        errors="coerce"
    )


# =====================================================================
# REMOVE ROWS WITH MISSING TARGET
# =====================================================================

rows_before_cleaning = len(data)

data = data.dropna(
    subset=[TARGET_COLUMN]
).reset_index(drop=True)

removed_rows = (
    rows_before_cleaning
    - len(data)
)

print(
    "\nRows removed due to missing target:",
    removed_rows
)

print(
    "Rows remaining:",
    len(data)
)

if len(data) < 10:

    raise ValueError(
        "Too few usable rows are available for training."
    )


# =====================================================================
# CREATE INPUT AND TARGET
# =====================================================================

X = data[
    feature_columns
].copy()

y = data[
    TARGET_COLUMN
].astype(float).copy()

print("\nInput shape:", X.shape)
print("Target shape:", y.shape)

print("\nTarget statistics:")
print(y.describe())


# =====================================================================
# TRAIN-TEST SPLIT
# =====================================================================

(
    X_train,
    X_test,
    y_train,
    y_test
) = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE
)

print("\nTraining records:", len(X_train))
print("Testing records:", len(X_test))


# =====================================================================
# PREPROCESSING
# =====================================================================

numeric_pipeline = Pipeline(
    steps=[
        (
            "numeric_imputer",
            SimpleImputer(
                strategy="median"
            )
        ),

        (
            "numeric_scaler",
            StandardScaler()
        )
    ]
)

categorical_pipeline = Pipeline(
    steps=[
        (
            "categorical_imputer",
            SimpleImputer(
                strategy="most_frequent"
            )
        ),

        (
            "one_hot_encoder",
            create_one_hot_encoder()
        )
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        (
            "numeric_features",
            numeric_pipeline,
            rainfall_columns
        ),

        (
            "categorical_features",
            categorical_pipeline,
            categorical_columns
        )
    ],

    remainder="drop"
)


# =====================================================================
# TRANSFORM DATA
# =====================================================================

X_train_processed = preprocessor.fit_transform(
    X_train
)

X_test_processed = preprocessor.transform(
    X_test
)

X_all_processed = preprocessor.transform(
    X
)

X_train_processed = np.asarray(
    X_train_processed,
    dtype=np.float32
)

X_test_processed = np.asarray(
    X_test_processed,
    dtype=np.float32
)

X_all_processed = np.asarray(
    X_all_processed,
    dtype=np.float32
)

y_train_array = np.asarray(
    y_train,
    dtype=np.float32
).reshape(-1)

y_test_array = np.asarray(
    y_test,
    dtype=np.float32
).reshape(-1)

print(
    "\nProcessed training shape:",
    X_train_processed.shape
)

print(
    "Processed testing shape:",
    X_test_processed.shape
)


# =====================================================================
# SAVE PREPROCESSOR
# =====================================================================

joblib.dump(
    preprocessor,
    AIS_PREPROCESSOR_PATH
)

print("\nAIS preprocessor saved:")
print(AIS_PREPROCESSOR_PATH)


# =====================================================================
# RUN AIS OPTIMIZATION
# =====================================================================

(
    best_ais_parameters,
    best_ais_cv_rmse,
    ais_optimization_history
) = artificial_immune_system_optimization(
    X_train_processed,
    y_train_array
)

ais_optimization_history.to_csv(
    AIS_OPTIMIZATION_CSV_PATH,
    index=False
)

print("\nAIS optimization history saved:")
print(AIS_OPTIMIZATION_CSV_PATH)

print("\nBest AIS parameters:")
print(best_ais_parameters)

print(
    "\nBest AIS cross-validation RMSE:",
    round(
        best_ais_cv_rmse,
        4
    )
)


# =====================================================================
# TRAIN AIS-OPTIMIZED RANDOM FOREST
# =====================================================================

print("\n" + "=" * 75)
print("TRAINING AIS-OPTIMIZED RANDOM FOREST")
print("=" * 75)

ais_model = RandomForestRegressor(
    n_estimators=best_ais_parameters[
        "n_estimators"
    ],

    max_depth=best_ais_parameters[
        "max_depth"
    ],

    min_samples_split=best_ais_parameters[
        "min_samples_split"
    ],

    min_samples_leaf=best_ais_parameters[
        "min_samples_leaf"
    ],

    max_features=best_ais_parameters[
        "max_features"
    ],

    bootstrap=best_ais_parameters[
        "bootstrap"
    ],

    random_state=RANDOM_STATE,

    n_jobs=1
)

ais_model.fit(
    X_train_processed,
    y_train_array
)

ais_predictions = ais_model.predict(
    X_test_processed
)

ais_predictions = np.maximum(
    ais_predictions,
    0
)

ais_metrics = calculate_metrics(
    y_test_array,
    ais_predictions
)

print("\nAIS Random Forest results:")

for metric_name, metric_value in ais_metrics.items():

    print(
        f"  {metric_name}: "
        f"{metric_value:.4f}"
    )


# =====================================================================
# TRAIN BASELINE MODELS FOR COMPARISON
# =====================================================================

comparison_models = {
    "Linear Regression": LinearRegression(),

    "Gradient Boosting": GradientBoostingRegressor(
        n_estimators=300,
        learning_rate=0.03,
        max_depth=3,
        random_state=RANDOM_STATE
    ),

    "Extra Trees": ExtraTreesRegressor(
        n_estimators=400,
        max_depth=10,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features="sqrt",
        random_state=RANDOM_STATE,
        n_jobs=1
    ),

    "Standard Random Forest": RandomForestRegressor(
        n_estimators=300,
        max_depth=10,
        min_samples_split=2,
        min_samples_leaf=1,
        max_features="sqrt",
        random_state=RANDOM_STATE,
        n_jobs=1
    )
}

model_metrics = {
    "AIS Random Forest": ais_metrics
}

model_predictions = {
    "AIS Random Forest": ais_predictions
}

print("\n" + "=" * 75)
print("TRAINING COMPARISON MODELS")
print("=" * 75)

for model_name, model in comparison_models.items():

    print(f"\nTraining {model_name}...")

    model.fit(
        X_train_processed,
        y_train_array
    )

    predictions = model.predict(
        X_test_processed
    )

    predictions = np.maximum(
        predictions,
        0
    )

    metrics = calculate_metrics(
        y_test_array,
        predictions
    )

    model_predictions[
        model_name
    ] = predictions

    model_metrics[
        model_name
    ] = metrics

    print(f"{model_name} results:")

    for metric_name, metric_value in metrics.items():

        print(
            f"  {metric_name}: "
            f"{metric_value:.4f}"
        )


# =====================================================================
# TRAIN ANN MODEL FOR H5 OUTPUT
# =====================================================================

print("\n" + "=" * 75)
print("TRAINING ARTIFICIAL NEURAL NETWORK")
print("=" * 75)

ann_model = create_ann_model(
    input_dimension=X_train_processed.shape[1]
)

ann_model.summary()

early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=35,
    restore_best_weights=True,
    verbose=1
)

reduce_learning_rate = ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=12,
    min_lr=0.00001,
    verbose=1
)

ann_history = ann_model.fit(
    X_train_processed,
    y_train_array,
    validation_split=0.20,
    epochs=ANN_EPOCHS,
    batch_size=ANN_BATCH_SIZE,
    callbacks=[
        early_stopping,
        reduce_learning_rate
    ],
    verbose=1
)

ann_predictions = ann_model.predict(
    X_test_processed,
    verbose=0
).reshape(-1)

ann_predictions = np.maximum(
    ann_predictions,
    0
)

ann_metrics = calculate_metrics(
    y_test_array,
    ann_predictions
)

model_predictions[
    "ANN"
] = ann_predictions

model_metrics[
    "ANN"
] = ann_metrics

print("\nANN results:")

for metric_name, metric_value in ann_metrics.items():

    print(
        f"  {metric_name}: "
        f"{metric_value:.4f}"
    )


# =====================================================================
# SAVE ANN MODEL
# =====================================================================

ann_model.save(
    AIS_H5_MODEL_PATH
)

print("\nAIS ANN model saved:")
print(AIS_H5_MODEL_PATH)


# =====================================================================
# SAVE AIS MODEL PACKAGE
# =====================================================================

ais_model_package = {
    "project_name": (
        "Smart Rainfall Pattern Analysis and "
        "Localized Rainfall Prediction System"
    ),

    "optimization_method": (
        "Artificial Immune System Clonal Selection"
    ),

    "model_name": (
        "AIS-Optimized Random Forest Regressor"
    ),

    "model": ais_model,

    "preprocessor": preprocessor,

    "feature_columns": feature_columns,

    "categorical_columns": categorical_columns,

    "rainfall_columns": rainfall_columns,

    "target_column": TARGET_COLUMN,

    "best_parameters": best_ais_parameters,

    "cross_validation_rmse": best_ais_cv_rmse,

    "test_metrics": ais_metrics,

    "created_at": datetime.now().isoformat()
}

joblib.dump(
    ais_model_package,
    AIS_PKL_MODEL_PATH
)

print("\nAIS PKL model saved:")
print(AIS_PKL_MODEL_PATH)


# =====================================================================
# CREATE AIS RESULT CSV
# =====================================================================

ais_result_data = (
    X_test[categorical_columns]
    .copy()
    .reset_index(drop=True)
)

ais_result_data[
    "Actual_Rainfall_11_Sep"
] = np.round(
    y_test_array,
    4
)

ais_result_data[
    "AIS_Predicted_Rainfall_11_Sep"
] = np.round(
    ais_predictions,
    4
)

ais_result_data[
    "Absolute_Error"
] = np.round(
    np.abs(
        y_test_array
        - ais_predictions
    ),
    4
)

ais_result_data[
    "Squared_Error"
] = np.round(
    (
        y_test_array
        - ais_predictions
    ) ** 2,
    4
)

ais_result_data[
    "Prediction_Difference"
] = np.round(
    ais_predictions
    - y_test_array,
    4
)

ais_result_data[
    "Model"
] = "AIS Random Forest"

ais_result_data.to_csv(
    AIS_RESULT_CSV_PATH,
    index=False
)

print("\nAIS result CSV saved:")
print(AIS_RESULT_CSV_PATH)

print("\nAIS result preview:")
print(ais_result_data.head())


# =====================================================================
# CREATE AIS PREDICTION CSV FOR ALL LOCATIONS
# =====================================================================

all_ais_predictions = ais_model.predict(
    X_all_processed
)

all_ais_predictions = np.maximum(
    all_ais_predictions,
    0
)

ais_prediction_data = (
    data[categorical_columns]
    .copy()
    .reset_index(drop=True)
)

ais_prediction_data[
    "Actual_Rainfall_11_Sep"
] = np.round(
    y.to_numpy(),
    4
)

ais_prediction_data[
    "AIS_Predicted_Rainfall_11_Sep"
] = np.round(
    all_ais_predictions,
    4
)

ais_prediction_data[
    "Absolute_Error"
] = np.round(
    np.abs(
        y.to_numpy()
        - all_ais_predictions
    ),
    4
)

ais_prediction_data[
    "Rainfall_Category"
] = np.where(
    ais_prediction_data[
        "AIS_Predicted_Rainfall_11_Sep"
    ] >= 64.5,

    "Heavy Rainfall",

    np.where(
        ais_prediction_data[
            "AIS_Predicted_Rainfall_11_Sep"
        ] >= 15.6,

        "Moderate Rainfall",

        np.where(
            ais_prediction_data[
                "AIS_Predicted_Rainfall_11_Sep"
            ] >= 2.5,

            "Light Rainfall",

            "Very Light or No Rainfall"
        )
    )
)

ais_prediction_data[
    "Model_Used"
] = "AIS-Optimized Random Forest"

ais_prediction_data.to_csv(
    AIS_PREDICTION_CSV_PATH,
    index=False
)

print("\nAIS prediction CSV saved:")
print(AIS_PREDICTION_CSV_PATH)

print("\nAIS prediction preview:")
print(ais_prediction_data.head())


# =====================================================================
# AIS ACCURACY GRAPH
# =====================================================================

model_names = list(
    model_metrics.keys()
)

r2_scores = [
    model_metrics[model_name][
        "R2_Score"
    ]
    for model_name in model_names
]

plt.figure(
    figsize=(13, 7)
)

bars = plt.bar(
    model_names,
    r2_scores
)

plt.axhline(
    y=0,
    linewidth=1
)

plt.title(
    "AIS Rainfall Prediction Model Accuracy",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Regression models",
    fontsize=12
)

plt.ylabel(
    "R² score",
    fontsize=12
)

plt.xticks(
    rotation=20
)

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

for bar, score in zip(
    bars,
    r2_scores
):

    if score >= 0:

        text_y = score + 0.02
        vertical_alignment = "bottom"

    else:

        text_y = score - 0.04
        vertical_alignment = "top"

    plt.text(
        bar.get_x()
        + bar.get_width() / 2,
        text_y,
        f"{score:.3f}",
        ha="center",
        va=vertical_alignment,
        fontweight="bold"
    )

save_and_show_graph(
    AIS_ACCURACY_GRAPH_PATH
)


# =====================================================================
# AIS MODEL COMPARISON GRAPH
# =====================================================================

mae_values = [
    model_metrics[model_name]["MAE"]
    for model_name in model_names
]

rmse_values = [
    model_metrics[model_name]["RMSE"]
    for model_name in model_names
]

positions = np.arange(
    len(model_names)
)

bar_width = 0.35

plt.figure(
    figsize=(14, 7)
)

mae_bars = plt.bar(
    positions - bar_width / 2,
    mae_values,
    width=bar_width,
    label="MAE"
)

rmse_bars = plt.bar(
    positions + bar_width / 2,
    rmse_values,
    width=bar_width,
    label="RMSE"
)

plt.title(
    "AIS Model Error Comparison",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Regression models",
    fontsize=12
)

plt.ylabel(
    "Error value",
    fontsize=12
)

plt.xticks(
    positions,
    model_names,
    rotation=20
)

plt.legend()

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.4
)

for bar_collection in [
    mae_bars,
    rmse_bars
]:

    for bar in bar_collection:

        height = bar.get_height()

        plt.text(
            bar.get_x()
            + bar.get_width() / 2,
            height,
            f"{height:.2f}",
            ha="center",
            va="bottom",
            fontsize=8
        )

save_and_show_graph(
    AIS_COMPARISON_GRAPH_PATH
)


# =====================================================================
# AIS RAINFALL CORRELATION HEATMAP
# =====================================================================

heatmap_columns = (
    rainfall_columns
    + [TARGET_COLUMN]
)

correlation_matrix = (
    data[heatmap_columns]
    .corr()
)

plt.figure(
    figsize=(20, 16)
)

heatmap_image = plt.imshow(
    correlation_matrix,
    aspect="auto",
    cmap="coolwarm",
    vmin=-1,
    vmax=1
)

plt.colorbar(
    heatmap_image,
    label="Correlation coefficient"
)

plt.title(
    "AIS Rainfall Correlation Heatmap",
    fontsize=18,
    fontweight="bold"
)

plt.xticks(
    ticks=np.arange(
        len(correlation_matrix.columns)
    ),
    labels=correlation_matrix.columns,
    rotation=90,
    fontsize=7
)

plt.yticks(
    ticks=np.arange(
        len(correlation_matrix.index)
    ),
    labels=correlation_matrix.index,
    fontsize=7
)

save_and_show_graph(
    AIS_HEATMAP_PATH
)


# =====================================================================
# AIS RESULT GRAPH
# =====================================================================

sorted_result_data = (
    ais_result_data
    .sort_values(
        by="Actual_Rainfall_11_Sep"
    )
    .reset_index(drop=True)
)

result_sample_numbers = np.arange(
    1,
    len(sorted_result_data) + 1
)

plt.figure(
    figsize=(14, 7)
)

plt.plot(
    result_sample_numbers,
    sorted_result_data[
        "Actual_Rainfall_11_Sep"
    ],
    marker="o",
    linewidth=2,
    label="Actual rainfall"
)

plt.plot(
    result_sample_numbers,
    sorted_result_data[
        "AIS_Predicted_Rainfall_11_Sep"
    ],
    marker="s",
    linewidth=2,
    label="AIS predicted rainfall"
)

plt.title(
    "AIS Actual versus Predicted Rainfall Results",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Test sample number",
    fontsize=12
)

plt.ylabel(
    "Rainfall in millimetres",
    fontsize=12
)

plt.xticks(
    result_sample_numbers
)

plt.legend()

plt.grid(
    linestyle="--",
    alpha=0.4
)

save_and_show_graph(
    AIS_RESULT_GRAPH_PATH
)


# =====================================================================
# AIS PREDICTION GRAPH
# =====================================================================

sorted_prediction_data = (
    ais_prediction_data
    .sort_values(
        by="Actual_Rainfall_11_Sep"
    )
    .reset_index(drop=True)
)

location_numbers = np.arange(
    1,
    len(sorted_prediction_data) + 1
)

plt.figure(
    figsize=(16, 8)
)

plt.plot(
    location_numbers,
    sorted_prediction_data[
        "Actual_Rainfall_11_Sep"
    ],
    marker="o",
    markersize=4,
    linewidth=1.8,
    label="Actual rainfall"
)

plt.plot(
    location_numbers,
    sorted_prediction_data[
        "AIS_Predicted_Rainfall_11_Sep"
    ],
    marker="s",
    markersize=4,
    linewidth=1.8,
    label="AIS predicted rainfall"
)

plt.title(
    "AIS Localized Rainfall Prediction for All Locations",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Location number",
    fontsize=12
)

plt.ylabel(
    "Rainfall in millimetres",
    fontsize=12
)

plt.legend()

plt.grid(
    linestyle="--",
    alpha=0.4
)

save_and_show_graph(
    AIS_PREDICTION_GRAPH_PATH
)


# =====================================================================
# AIS OPTIMIZATION GRAPH
# =====================================================================

plt.figure(
    figsize=(12, 7)
)

plt.plot(
    ais_optimization_history[
        "Generation"
    ],
    ais_optimization_history[
        "Best_CV_RMSE"
    ],
    marker="o",
    linewidth=2,
    label="Global best CV RMSE"
)

plt.plot(
    ais_optimization_history[
        "Generation"
    ],
    ais_optimization_history[
        "Average_CV_RMSE"
    ],
    marker="s",
    linewidth=2,
    label="Population average CV RMSE"
)

plt.title(
    "Artificial Immune System Optimization Progress",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "AIS generation",
    fontsize=12
)

plt.ylabel(
    "Cross-validation RMSE",
    fontsize=12
)

plt.xticks(
    ais_optimization_history[
        "Generation"
    ]
)

plt.legend()

plt.grid(
    linestyle="--",
    alpha=0.4
)

save_and_show_graph(
    AIS_OPTIMIZATION_GRAPH_PATH
)


# =====================================================================
# ANN TRAINING HISTORY GRAPH
# =====================================================================

ann_training_loss = ann_history.history.get(
    "loss",
    []
)

ann_validation_loss = ann_history.history.get(
    "val_loss",
    []
)

epochs_completed = range(
    1,
    len(ann_training_loss) + 1
)

plt.figure(
    figsize=(12, 7)
)

plt.plot(
    epochs_completed,
    ann_training_loss,
    linewidth=2,
    label="Training loss"
)

plt.plot(
    epochs_completed,
    ann_validation_loss,
    linewidth=2,
    label="Validation loss"
)

plt.title(
    "AIS Project ANN Training History",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Epoch",
    fontsize=12
)

plt.ylabel(
    "Mean squared error loss",
    fontsize=12
)

plt.legend()

plt.grid(
    linestyle="--",
    alpha=0.4
)

save_and_show_graph(
    AIS_ANN_HISTORY_GRAPH_PATH
)


# =====================================================================
# FEATURE IMPORTANCE GRAPH
# =====================================================================

try:

    processed_feature_names = (
        preprocessor
        .get_feature_names_out()
    )

except Exception:

    processed_feature_names = np.array(
        [
            f"Feature_{index + 1}"
            for index in range(
                X_train_processed.shape[1]
            )
        ]
    )

feature_importance_data = pd.DataFrame(
    {
        "Feature": processed_feature_names,

        "Importance": (
            ais_model.feature_importances_
        )
    }
)

feature_importance_data = (
    feature_importance_data
    .sort_values(
        by="Importance",
        ascending=False
    )
    .head(20)
    .sort_values(
        by="Importance",
        ascending=True
    )
)

plt.figure(
    figsize=(12, 9)
)

plt.barh(
    feature_importance_data[
        "Feature"
    ],
    feature_importance_data[
        "Importance"
    ]
)

plt.title(
    "Top AIS Random Forest Feature Importances",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel(
    "Feature importance",
    fontsize=12
)

plt.ylabel(
    "Feature",
    fontsize=12
)

plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.4
)

save_and_show_graph(
    AIS_FEATURE_IMPORTANCE_PATH
)


# =====================================================================
# SAVE AIS YAML CONFIGURATION
# =====================================================================

yaml_data = {
    "project": {
        "name": (
            "Smart Rainfall Pattern Analysis and "
            "Localized Rainfall Prediction System"
        ),

        "method": (
            "Artificial Immune System "
            "Clonal Selection Algorithm"
        ),

        "created_at": datetime.now().isoformat()
    },

    "dataset": {
        "path": DATASET_PATH,
        "rows": int(data.shape[0]),
        "columns": int(data.shape[1]),
        "training_rows": int(len(X_train)),
        "testing_rows": int(len(X_test)),
        "target_column": TARGET_COLUMN
    },

    "features": {
        "categorical_columns": categorical_columns,
        "rainfall_columns": rainfall_columns,

        "processed_feature_count": int(
            X_train_processed.shape[1]
        )
    },

    "ais_settings": {
        "population_size": AIS_POPULATION_SIZE,
        "generations": AIS_GENERATIONS,

        "selected_antibodies": (
            AIS_SELECTED_ANTIBODIES
        ),

        "clones_per_antibody": (
            AIS_CLONES_PER_ANTIBODY
        ),

        "random_replacements": (
            AIS_RANDOM_REPLACEMENTS
        )
    },

    "ais_best_parameters": (
        best_ais_parameters
    ),

    "ais_best_cv_rmse": float(
        best_ais_cv_rmse
    ),

    "ais_test_metrics": ais_metrics,

    "all_model_metrics": model_metrics,

    "saved_files": {
        "h5_model": AIS_H5_MODEL_PATH,
        "pkl_model": AIS_PKL_MODEL_PATH,

        "preprocessor": (
            AIS_PREPROCESSOR_PATH
        ),

        "yaml_configuration": AIS_YAML_PATH,
        "json_results": AIS_JSON_PATH,

        "result_csv": (
            AIS_RESULT_CSV_PATH
        ),

        "prediction_csv": (
            AIS_PREDICTION_CSV_PATH
        ),

        "optimization_csv": (
            AIS_OPTIMIZATION_CSV_PATH
        ),

        "accuracy_graph": (
            AIS_ACCURACY_GRAPH_PATH
        ),

        "comparison_graph": (
            AIS_COMPARISON_GRAPH_PATH
        ),

        "heatmap": AIS_HEATMAP_PATH,

        "result_graph": (
            AIS_RESULT_GRAPH_PATH
        ),

        "prediction_graph": (
            AIS_PREDICTION_GRAPH_PATH
        ),

        "optimization_graph": (
            AIS_OPTIMIZATION_GRAPH_PATH
        ),

        "ann_history_graph": (
            AIS_ANN_HISTORY_GRAPH_PATH
        ),

        "feature_importance_graph": (
            AIS_FEATURE_IMPORTANCE_PATH
        )
    }
}

with open(
    AIS_YAML_PATH,
    "w",
    encoding="utf-8"
) as yaml_file:

    yaml.safe_dump(
        yaml_data,
        yaml_file,
        sort_keys=False,
        allow_unicode=True
    )

print("\nAIS YAML configuration saved:")
print(AIS_YAML_PATH)


# =====================================================================
# SAVE AIS JSON RESULTS
# =====================================================================

json_data = {
    "project_name": (
        "Smart Rainfall Pattern Analysis and "
        "Localized Rainfall Prediction System"
    ),

    "optimization_method": (
        "Artificial Immune System "
        "Clonal Selection Algorithm"
    ),

    "created_at": datetime.now().isoformat(),

    "dataset_information": {
        "dataset_path": DATASET_PATH,

        "number_of_rows": int(
            data.shape[0]
        ),

        "number_of_columns": int(
            data.shape[1]
        ),

        "training_records": int(
            len(X_train)
        ),

        "testing_records": int(
            len(X_test)
        ),

        "target_column": TARGET_COLUMN
    },

    "ais_configuration": {
        "population_size": AIS_POPULATION_SIZE,
        "generations": AIS_GENERATIONS,

        "selected_antibodies": (
            AIS_SELECTED_ANTIBODIES
        ),

        "clones_per_antibody": (
            AIS_CLONES_PER_ANTIBODY
        ),

        "random_replacements": (
            AIS_RANDOM_REPLACEMENTS
        )
    },

    "best_ais_parameters": (
        best_ais_parameters
    ),

    "best_cross_validation_rmse": float(
        best_ais_cv_rmse
    ),

    "ais_test_metrics": ais_metrics,

    "all_model_results": model_metrics,

    "target_statistics": {
        "minimum": float(y.min()),
        "maximum": float(y.max()),
        "mean": float(y.mean()),
        "median": float(y.median()),

        "standard_deviation": float(
            y.std()
        )
    },

    "ann_training": {
        "epochs_completed": int(
            len(ann_training_loss)
        ),

        "final_training_loss": (
            float(ann_training_loss[-1])
            if ann_training_loss
            else None
        ),

        "final_validation_loss": (
            float(ann_validation_loss[-1])
            if ann_validation_loss
            else None
        )
    },

    "saved_files": {
        "h5_model": AIS_H5_MODEL_PATH,
        "pkl_model": AIS_PKL_MODEL_PATH,

        "preprocessor": (
            AIS_PREPROCESSOR_PATH
        ),

        "yaml_configuration": AIS_YAML_PATH,
        "json_results": AIS_JSON_PATH,

        "result_csv": (
            AIS_RESULT_CSV_PATH
        ),

        "prediction_csv": (
            AIS_PREDICTION_CSV_PATH
        ),

        "optimization_csv": (
            AIS_OPTIMIZATION_CSV_PATH
        ),

        "accuracy_graph": (
            AIS_ACCURACY_GRAPH_PATH
        ),

        "comparison_graph": (
            AIS_COMPARISON_GRAPH_PATH
        ),

        "heatmap": AIS_HEATMAP_PATH,

        "result_graph": (
            AIS_RESULT_GRAPH_PATH
        ),

        "prediction_graph": (
            AIS_PREDICTION_GRAPH_PATH
        ),

        "optimization_graph": (
            AIS_OPTIMIZATION_GRAPH_PATH
        ),

        "ann_history_graph": (
            AIS_ANN_HISTORY_GRAPH_PATH
        ),

        "feature_importance_graph": (
            AIS_FEATURE_IMPORTANCE_PATH
        )
    }
}

with open(
    AIS_JSON_PATH,
    "w",
    encoding="utf-8"
) as json_file:

    json.dump(
        json_data,
        json_file,
        indent=4,
        ensure_ascii=False,
        default=json_converter
    )

print("\nAIS JSON results saved:")
print(AIS_JSON_PATH)


# =====================================================================
# VERIFY SAVED AIS MODEL
# =====================================================================

print("\n" + "=" * 75)
print("VERIFYING SAVED AIS MODEL")
print("=" * 75)

loaded_package = joblib.load(
    AIS_PKL_MODEL_PATH
)

loaded_model = loaded_package[
    "model"
]

loaded_preprocessor = loaded_package[
    "preprocessor"
]

sample_input = X.iloc[
    [0]
].copy()

sample_processed = loaded_preprocessor.transform(
    sample_input
)

sample_processed = np.asarray(
    sample_processed,
    dtype=np.float32
)

sample_prediction = loaded_model.predict(
    sample_processed
)[0]

sample_prediction = max(
    float(sample_prediction),
    0
)

print("\nSample location:")

for column in categorical_columns:

    print(
        f"{column}: "
        f"{sample_input.iloc[0][column]}"
    )

print(
    "\nAIS predicted rainfall for 11-Sep:",
    round(
        sample_prediction,
        2
    ),
    "mm"
)

print(
    "Actual rainfall for 11-Sep:",
    round(
        float(y.iloc[0]),
        2
    ),
    "mm"
)


# =====================================================================
# FINAL SUMMARY
# =====================================================================

print("\n" + "=" * 75)
print("AIS PROJECT COMPLETED SUCCESSFULLY")
print("=" * 75)

generated_files = [
    AIS_H5_MODEL_PATH,
    AIS_PKL_MODEL_PATH,
    AIS_PREPROCESSOR_PATH,
    AIS_YAML_PATH,
    AIS_JSON_PATH,
    AIS_RESULT_CSV_PATH,
    AIS_PREDICTION_CSV_PATH,
    AIS_OPTIMIZATION_CSV_PATH,
    AIS_ACCURACY_GRAPH_PATH,
    AIS_COMPARISON_GRAPH_PATH,
    AIS_HEATMAP_PATH,
    AIS_RESULT_GRAPH_PATH,
    AIS_PREDICTION_GRAPH_PATH,
    AIS_OPTIMIZATION_GRAPH_PATH,
    AIS_ANN_HISTORY_GRAPH_PATH,
    AIS_FEATURE_IMPORTANCE_PATH
]

print("\nGenerated AIS files:\n")

for file_path in generated_files:

    if os.path.exists(file_path):

        file_size = os.path.getsize(
            file_path
        )

        print(
            "[CREATED]",
            os.path.basename(file_path),
            f"({file_size:,} bytes)"
        )

    else:

        print(
            "[NOT CREATED]",
            os.path.basename(file_path)
        )


print("\nAIS optimized parameters:")

for parameter_name, parameter_value in (
    best_ais_parameters.items()
):

    print(
        f"  {parameter_name}: "
        f"{parameter_value}"
    )


print("\nAIS Random Forest performance:")

print(
    "  MAE:",
    round(
        ais_metrics["MAE"],
        4
    )
)

print(
    "  MSE:",
    round(
        ais_metrics["MSE"],
        4
    )
)

print(
    "  RMSE:",
    round(
        ais_metrics["RMSE"],
        4
    )
)

print(
    "  R² Score:",
    round(
        ais_metrics["R2_Score"],
        4
    )
)

print(
    "\nAll AIS files were saved in:\n",
    PROJECT_FOLDER
)
```
